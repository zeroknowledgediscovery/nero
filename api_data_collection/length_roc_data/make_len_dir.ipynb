```python
import os

length = 1000

in_dir = f"{os.getcwd()}\\150k\\text"

out_dir = "1k"
out_dir = f"{os.getcwd()}\\{out_dir}"
text_dir = f"{out_dir}\\text"
rates_dir = f"{out_dir}\\rates"
if not os.path.exists(out_dir):
    os.mkdir(out_dir)
    os.mkdir(text_dir)
    os.mkdir(rates_dir)


for file in os.listdir(in_dir):
    out_text = ""
    with open(f"{in_dir}\\{file}","r",encoding ="utf8") as infile:
        text = infile.read()
        middle_idx = int(len(text)/2)
        begin_idx = middle_idx - (length//2)
        end_idx = middle_idx + (length//2)
        out_text = text[begin_idx:end_idx]
        print(f"{out_text}\n")
        with open(f"{text_dir}\\{file}","w",encoding = "utf8") as outfile:
            outfile.write(out_text)

```
